```python
# extracting the text from the pdf
!pip install PyPDF2
import PyPDF2
import os
# read read pdf from the link
doc = PyPDF2.PdfReader("Attention all you need.pdf")
# extract the text from the pdf
text = ""
for page in doc.pages:
    text += page.extract_text()
    print(text)
```

```python
# make the text into chunks of sentences
sentences = text.split('. ')
sentences
```

```python
# prompt: define embeddings function and pass the sentence chunks into the embeddings

def embeddings(text, model_name='sentence-transformers/all-MiniLM-L6-v2'):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1)
        return embeddings.numpy()

sentence_embeddings = [embeddings(sentence) for sentence in sentences]
```

```python
sentence_embeddings[0]
```

```python
query="Encoder and Decoder stacks"
query_embedding=embeddings(query)
```

```python
# prform similarity search
from sklearn.metrics.pairwise import cosine_similarity
import torch

# Convert query_embedding to a PyTorch tensor
query_embedding_tensor = torch.from_numpy(query_embedding)

# Convert sentence_embeddings to a PyTorch tensor
sentence_embeddings_tensor = torch.stack([torch.from_numpy(embedding) for embedding in sentence_embeddings])

# Calculate cosine similarity
similarities = torch.cosine_similarity(query_embedding_tensor, sentence_embeddings_tensor)

# Find the index of the most similar sentence
most_similar_index = torch.argmax(similarities)

# Get the similarity score of the most similar sentence
most_similar_score = similarities[most_similar_index]

# Print the results
print(f"Query: {query}")
print(f"Most similar document: {sentence_embeddings[most_similar_index]}")
```
