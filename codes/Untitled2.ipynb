```python
import pandas as pd
import numpy as np

# Try reading the file with a different encoding, such as 'latin-1'
df = pd.read_csv('Data.csv', encoding='ISO-8859-1')

df
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]","",regex=True,inplace=True)
data
```

```python
for i in range(25):
    data.iloc[:,i]=data.iloc[:,i].str.lower()
```

```python
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data

```

```python
# extracting the text of all headlnes and storing in array
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
headlines[25]
```

```python
# bag of words creation for the headlines data
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(2,2))
X=cv.fit_transform(headlines)
```

```python
# apply random forest
import sklearn
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier(n_estimators=200,criterion='entropy')
random_forest.fit(X,train['Label'])
```

```python
test_headlines=[]

for row in range(0,len(test.index)):
    test_headlines.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_headlines[1]
```

```python
test_dataset=cv.transform(test_headlines)
predictions=random_forest.predict(test_dataset)
```

```python
# accuray
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
```

```python
# support vector machine
from sklearn.svm import SVC
svc=SVC()
svc.fit(X,train['Label'])
predictions=svc.predict(test_dataset)
```

```python
# accuracy prediction
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
```
