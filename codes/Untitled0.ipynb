```python
import tkinter as tk
from tkinter import messagebox

def register():
    username = entry_username.get()
    email = entry_email.get()
    password = entry_password.get()
    confirm_password = entry_confirm_password.get()

    if not username or not email or not password or not confirm_password:
        messagebox.showerror("Error", "All fields are required.")
        return

    if password != confirm_password:
        messagebox.showerror("Error", "Passwords do not match.")
        return

    # Here you would add your logic to register the user with the provided username, email, and password

    messagebox.showinfo("Success", "User registered successfully.")

    root = tk.Tk()
    root.title("Registration")

# Frame for registration fields
frame_registration = tk.Frame(root)
frame_registration.pack(padx=10, pady=10)

tk.Label(frame_registration, text="Username:").grid(row=0, column=0, padx=(0, 10))
entry_username = tk.Entry(frame_registration)
entry_username.grid(row=0, column=1)

tk.Label(frame_registration, text="Email:").grid(row=1, column=0, padx=(0, 10))
entry_email = tk.Entry(frame_registration)
entry_email.grid(row=1, column=1)

tk.Label(frame_registration, text="Password:").grid(row=2, column=0, padx=(0, 10))
entry_password = tk.Entry(frame_registration, show='*')
entry_password.grid(row=2, column=1)

tk.Label(frame_registration, text="Confirm Password:").grid(row=3, column=0, padx=(0, 10))
entry_confirm_password = tk.Entry(frame_registration, show='*')
entry_confirm_password.grid(row=3, column=1)

# Frame for buttons
frame_buttons = tk.Frame(root)
frame_buttons.pack(pady=10)

button_register = tk.Button(frame_buttons, text="Register", command=register)
button_register.grid(row=0, column=0, padx=(0, 10))

button_login = tk.Button(frame_buttons, text="Login")
button_login.grid(row=0, column=1)

root.mainloop()
```
