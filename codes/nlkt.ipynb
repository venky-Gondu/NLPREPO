```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import sent_tokenize
import re

# Make sure NLTK resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

ps = PorterStemmer()
wordnet = WordNetLemmatizer()

sentences = sent_tokenize(paragraph)
print("Tokenized sentences:", sentences)  # Debug print

corpus = []

for sentence in sentences:
    print("Processing sentence:", sentence)  # Debug print
    words = re.sub('[^a-zA-Z]', ' ', sentence)
    words = words.lower()
    words = words.split()
    words = [ps.stem(word) for word in words if not word in set(stopwords.words('english'))]
    words = ' '.join(words)
    #print("Processed words:", words)  # Debug print
    corpus.append(words)

  # Debug print
```

```python
import pandas as pd
data =pd.read_csv('SMSSpamCollection', sep = '\t',
                      names=['label', 'message'])
data
```

```python
# remove stop words from the messages and make lower cases
import re
import nltk
nltk.download('stopwords')
for i in range(len(data)):
  review = re.sub('[^a-zA-Z]', ' ', data['message'][i])
  review = review.lower()
  review = review.split()
  review= [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  data['message'][i] = review
```

```python
data
```

```python
# convert the message into bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 3000)
X = cv.fit_transform(data['message']).toarray()
print(X)
```

```python
df= pd.DataFrame(X)
df
```

```python
# label the ham and spam into binary classification of 1 and 0
y = pd.get_dummies(data['label'], drop_first=True)
y = y.astype(int)  # Convert True/False to 1/0
#
```

```python
y
```

```python
# spliit the data into test and train set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 200)
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
predication = spam_detect_model.predict(X_test)
```

```python
# accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predication)
```

```python
# presenting confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predication)
```
